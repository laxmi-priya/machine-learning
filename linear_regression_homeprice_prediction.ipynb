```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sq ft)')
plt.ylabel('price($)')
plt.title('Rate chart')
plt.scatter(df.area,df.price,color='red',marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[1800]])
```

```python
dnf = pd.read_csv("area.csv")
p=reg.predict(dnf)
dnf['prices'] = p
```

```python
dnf.to_csv('prediction.csv',index=False)
```

```python
%matplotlib inline
plt.xlabel('area(sq ft)')
plt.ylabel('price($)')
plt.title('Rate chart Prediction')
plt.scatter(df.area,df.price,color='red',marker='*')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```
